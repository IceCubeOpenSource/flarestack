```python
%matplotlib inline
import matplotlib.pyplot as plt
from flarestack.data.simulate.simcube import simcube_dataset
sim_season = simcube_dataset.get_single_season("IC86-2012")
```

```python
sim_data = sim_season.get_background_model()
print(sim_data)
print(len(sim_data), "events in dataset")
```

```python
for key in ["logE", "sinDec", "time", "ra", "sigma"]:
    plt.figure()
    plt.hist(sim_data[key])
    plt.yscale("log")
    plt.title(key)
```
